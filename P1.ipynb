```python
from pandas import *
import matplotlib.pyplot as plt
from scipy import stats
set_option('display.max_rows', None)
set_option('display.max_columns', None)
```

```python
data_main = read_csv("C:/Users/Charles Lee/OneDrive - CUHK-Shenzhen/FIN3080_P3/data1.csv")
```

```python
data_main = data_main[data_main['code'] == 300]
data_main['date'] = to_datetime(data_main['date'])
```

```python
data_main = data_main.resample('ME', on='date').last()
data_main['return'] = data_main['close'].pct_change()
data_main
```

```python
returns = data_main['return'].dropna()
mean = returns.mean()
std_dev = returns.std()
skewness = returns.skew()
kurtosis = returns.kurtosis()
print(f'Mean: {mean}')
print(f'Standard Deviation: {std_dev}')
print(f'Skewness: {skewness}')
print(f'Kurtosis: {kurtosis}')
```

```python
plt.figure(figsize=(10, 6))
plt.title('CSI 300 Monthly Returns Histogram')
plt.xlabel('Returns')
plt.ylabel('Density')
plt.hist(returns, bins=50, alpha=1, color='blue', edgecolor='black', density=True)
plt.grid(True)
plt.savefig('return.png', dpi=300)
plt.show()
```

```python
statistic, p_value = stats.shapiro(returns)
print("Shapiro-Wilk Test Statistic:", statistic)
print("P-value:", p_value)
alpha = 0.05  # Significance level
if p_value > alpha:
    print("Sample looks Gaussian (fail to reject H0)")
else:
    print("Sample does not look Gaussian (reject H0)")
```
